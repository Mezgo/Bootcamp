```python
# 'dataset' contiene los datos de entrada para este script

# se importan las librerias necesarias para aplicar las funciones
import pandas as pd
import numpy as np

dfHotelBooking = pd.read_csv("hotel-bookings.csv")

```

```python
dfHotelBooking.head()
```

```python
dfHotelBooking = dfHotelBooking.drop(['adr'], axis = 1)


# Llenar los valores nulos
# los valores numericos se pueden llenar con la media
dfHotelBooking['children'].fillna((dfHotelBooking['children'].mean()), inplace=True)

# llenar valores faltantes con un valor específico
dfHotelBooking['company'].fillna(0, inplace=True)
dfHotelBooking['country'].fillna('No tiene', inplace=True)

# llenar con el valor del siguiente registro
dfHotelBooking['agent'].fillna(method="ffill", inplace=True)


# Se cambia el tipo de los datos
dfHotelBooking['children'] = dfHotelBooking['children'].values.astype(int)
dfHotelBooking['adults'] = dfHotelBooking['adults'].values.astype(int)
dfHotelBooking['babies'] = dfHotelBooking['babies'].values.astype(int)

# Se crea una nueva columna
# se crea la columna total de personas, la cual será la suma de la cantidad de cada grupo de persona
dfHotelBooking['Cant de personas'] = dfHotelBooking['babies'] + dfHotelBooking['children'] + dfHotelBooking['adults']

# Reemplazar un valor por otro
# El termino “TA” es “Travel Agents” y “TO” es “Tour Operators”
dfHotelBooking["market_segment"].replace({"Online TA": "Online Travel Agents", "Offline TA/TO": "Off Travel Agents/Tour Operators "}, inplace=True)

# Se toma los nombres de las columnas (encabezados) para pasarlos por el melt
data_top = list(dfHotelBooking.columns.values)

```

```python
dfHotelBooking.head()
```

```python
# Se elimina la columna adults, children y babies ya que estas son a las que se les realizará el melt
data_top.remove('adults')
data_top.remove('children')
data_top.remove('babies')
# se utiliza pandas.melt para reorganizar la estructura del dataframe
dfHotelBooking=dfHotelBooking.melt(data_top, var_name='Grupo Persona') 
```

```python
dfHotelBooking.head(10)
```

```python
#Se cambia el nombre de la columna 
dfHotelBooking= dfHotelBooking.rename(columns = {'value':'Cantidad'}, inplace = False)

# La función drop_duplicates() retorna un dataframe con las filas duplicadas eliminadas
dfHotelBooking.drop_duplicates(inplace=True)

# filtrar columnas, se dejan solo los valores que tienen 2015 en la columna de año de llegada
dfHotelBooking = dfHotelBooking.loc[dfHotelBooking['arrival_date_year'] == 2015] 
# se verifica que se realizo el filtro ahora hay menos columnas
dfHotelBooking.info()

# crear una columna con numeros o valores aleatorios, que se entregan en un array
dfHotelBooking['randNumCol'] = np.random.choice(['cool', 9, 20], dfHotelBooking.shape[0])

# Aplicando multiples funciones de agregación a un una agrupación de los datos hecha con group by
# se debe agrgar a la lista el valor númerico que utilizaran las funciones de agregación, en este caso es cantidad
dfHotelBooking = dfHotelBooking[['country','Grupo Persona','Cantidad']].groupby(['country','Grupo Persona']).agg(['mean','count'])
dfHotelBooking.head()
```

```python
#se organizan nuevamente las columnas, eliminando un nivel de indice y reseteandolo
# se elimina un nivel
dfHotelBooking.columns = dfHotelBooking.columns.droplevel()

# se reseta el indice
dfHotelBooking = dfHotelBooking.reset_index()


# se puede generar una columna con un ranking, el metodo es min por lo que el valor mas pequeño tendra el primer puesto
# se aplica por la columna mean
dfHotelBooking['rank'] = dfHotelBooking['mean'].rank(method = 'min')


# se pueden ordenan los valores con la función sort_values
dfHotelBooking.sort_values('rank',inplace = True)

# con la función sample se puede tomar una muestra del dataset, con una cantidad especifica de filas
dfHotelBooking = dfHotelBooking.sample(n=100)
dfHotelBooking.shape
# queda un dataset con 10 filas y 5 columnas
```
