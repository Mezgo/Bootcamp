````python
# <figure> to generate the space where to draw
fig = plt.figure()

# <axes> to command the type of draw
ax = plt.axes()

# Draw
ax.plot(x,y)
````

## **1.** **Line Plot**

### **Add two lines into the the same grafic**

````python
x = np.arange(0.0 , 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)

x2 = np.arange(0.0 , 2.0 , 0.01)
y2 = x2

fig = plt.figure()
ax  = plt.axes()

# Draw 2 times, so they overlap
ax.plot(x,y)
ax.plot(x2,y2)
````

### **Formating the line**

````python
fig = plt.figure()
ax  = plt.axes()

ax.plot(x, y, color= 'red', linewidth= 4, linestyle= 'dotted')
````

### **Figure & Axes Customization**

````python
ax = plt.axes()

ax.plot(x, y, color = 'k', linewidth = 2, linestyle = '-')

# Axes Labels
# <set> to add the lables and title of the figure
ax.set(xlabel= 'Time (s)',
       ylabel= 'Temperatutre (C)',
       title= 'Temperature Value')

# Margins
ax.set(xlim= (0,1), ylim= (0,4))

# Grid
ax.grid()
````

## **2.** **Legends**

````python
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)

x2 = np.arange(0.0, 2.0, 0.01)
y2 = x2

fig = plt.figure()
ax = plt.axes()

# Establish the names of each line in the label property.
ax.plot(x, y, label='Primera Curva')
ax.plot(x2,y2, label='Segunda Curva')

# Add the legend
ax.legend()
````

### Properties

````python
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)

x2 = np.arange(0.0, 2.0, 0.01)
y2 = x2

fig = plt.figure()
ax = plt.axes()

# Establish the names of each line in the label property.
ax.plot(x, y, label='Primera Curva')
ax.plot(x2,y2, label='Segunda Curva')

# Add the legend's properties
ax.legend(loc='upper center', shadow=True, fontsize=15)
````

## **3.** **Subplots**

This comes handy when the requierement is to show 2 graphics in the same figure. It needs to be defined as `subplots`.

This creates a `figure` element and several axes in the `axes` object.

    fig, ax = plt.subplots(n,m)

The `n` refers to the amount of ***rows***

The `m` refers to the amount of ***columns***


````python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

fig, axes = plt.subplots(2,1)
fig.suptitle('Vertically alligned graphs')

# 0 - 1 values reffers to the order of appereance.
axes[0].plot(x1, y1)
axes[1].plot(x2, y2)
````

For the 2 graphs share the `x` or `y` values, the following should be done:
``` python
    fig, axes = plt.subplots(2,1,sharex=True)
    # or 
    fig, axes = plt.subplots(2,1,sharey=True)
```
